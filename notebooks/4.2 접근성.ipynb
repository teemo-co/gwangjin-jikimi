## parking score 계산

```python
gdf = pd.read_parquet('../data/shp/100m_grids_with_metro_distances.parquet')
```

```python

child_speed = 1.0 # meters per second

scored_poi_df = calculate_parking_scores(gdf, merged, child_speed)
score_gdf = gpd.GeoDataFrame(scored_poi_df)

score_gdf['geometry'] = score_gdf.apply(
    lambda x: wkb.loads(x['geometry'], hex=True), axis=1
)
score_gdf.set_geometry('geometry', inplace=True)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 10))
score_gdf['parking_score'] = scaler.fit_transform(score_gdf[['parking_score']])

score_gdf[['gid','geometry','lng','lat','nearest_station','nearest_station_distance','parking_score']].to_parquet('../data/parking/grids_with_parking.parquet', index=False, engine='pyarrow')
score_gdf[['gid','nearest_station','nearest_station_distance','parking_score']].to_csv('../data/parking/parking_and_station.csv', index=False)
```

## 시각화

```python

# 파일 경로
shapefile_path ='../data/parking/grids_with_parking.parquet'
metro_path = "../data/metro/gwangjin_metro.csv"

# 1. Shapefile 불러오기 (좌표계 명시 후 변환)
gdf = gpd.read_parquet(shapefile_path)
gdf = gdf.set_crs(epsg=5179, allow_override=True).to_crs(epsg=4326)
gdf['original_parking_score'] = gdf['parking_score']
gdf['parking_score'] =gdf.parking_score.clip(0, gdf.parking_score.quantile(0.95))

# 2. 지하철 데이터 로딩
metro_df = pd.read_csv(metro_path)
metro_df = gpd.GeoDataFrame(
    metro_df,
    geometry=gpd.points_from_xy(metro_df.lng, metro_df.lat),
    crs="EPSG:4326" # 포인트 데이터의 CRS를 EPSG:4326으로 명시적 설정
)
metro_df = gpd.sjoin(metro_df, gdf, how="left", predicate='within')
```

```python
m = viz.initialize_map(gdf)
m = viz.add_circle_makers(m, metro_df, meters=500, group_name='Metro Points')
m = viz.add_heatmap(m, gdf, 'parking_score')
m
```
